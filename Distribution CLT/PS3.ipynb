    5. (7pt) Now create Rpairs of random realizations of X(i.e. sample size S=2). For each pair, compute its mean. You should have Rmeans. Make the histogram. How does this look like?
    Hint: while you can do this using loops, it is more useful to create a R×2 matrix of realizations of X, where each row represents one pair. Thereafter you compute means by rows and you have Rpair means. See python notes numpy statistical functions for an example.

```python
# Make a function to get the mean and the variance for later uses
def get_mean_var(x):
    mean_var = []
    mean_var.append(np.mean(x))
    return mean_var

# Make a function to get the R number of trails of size s of random drawing of 1 and -1
def get_random(s):
    Rpairs = []
    mean_Rpairs = []
    for i in range(0, R):
        current_trial = np.random.randint(0, 2, size = s) * 2 - 1
        Rpairs.append(current_trial)
        Rpairs_mean_var = get_mean_var(current_trial)
        mean_Rpairs.append(Rpairs_mean_var[0])
    return(mean_Rpairs)

# Create Rpairs of random realization of sample size 2 and calcultae mean and variance
mean_R_pairs = get_random(2)

# Plot the results
mean_plot = plt.hist(mean_R_pairs, bins = 100)
mean_plot = plt.title('Frquency of the Mean Values in R(1000) Trails')
mean_plot = plt.xlabel('Mean')
mean_plot = plt.ylabel('Count')               
```

From the histogram, we can see that roughly half of the trails out of R trials (1000) have a mean of 0. The rest have means of either -1 or 1. Within the trials with means of -1 and 1, approximately half has the mean of -1 and the other half with a mean of 1. The histogram may suggest a multimodal distribution.

    12. (8pt) Explain why do the histograms resemble normal distribution as Sgrows. 
    In particular, explain what happens when we move from single values S=1 to pairs S=2. Why did two equal peaks turn into a “山”-shaped histogram?

When S=1, our random variable X has a 0.5 probability of having a mean of -1 (tails) and a 0.5 probability of having a mean of 1 (heads). When we increase the size from 1 to 2, there are more combinations. We can have head-head, head-tail, tail-head, and tail-tail. In term of numbers, we will have (1, 1), (1, -1), (-1, 1), (-1, -1) and the means of those sets are 1, 0, 0, and -1. This explains why that extra bar is in the middle and makes the equal peaks into a 山 shaped histogram. As we increase the sizes, the combinations of heads and tails also increase and change the possible means, making the distribution more normal.
